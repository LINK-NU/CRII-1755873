```python
bid_data = pd.read_csv("../Data/bid_notick.txt", sep="|")
bid_data = bid_data[["ListingKey", "CreationDate", "MemberKey", "Bid_Key", "Amount", "ParticipationAmount", "ListingStatus", "Status"]]
bid_data = bid_data.rename(index=str, columns={"CreationDate": "BidCreationDate", "Amount": "BidAmount", "Bid_Key": "BidKey"})
listing_data = pd.read_csv('../Data/listing.txt', sep="|")
listing_data = listing_data[["Lst_Key", "CreationDate", "StartDate"]]
listing_data = listing_data.rename(index=str, columns={"Lst_Key": "ListingKey", "CreationDate": "ListingCreationDate", "StartDate": "ListingStartDate"})
data = bid_data.merge(listing_data, on="ListingKey", how="left")
data.sample(5)
```
