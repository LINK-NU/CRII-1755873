```python
years = np.arange(2002,2017)
months = np.arange(1,13)
dates = []
for year in years:
    for month in months:
        if month <10:
            dates.append(int(str(year)+"0"+str(month)))
        else:
            dates.append(int(str(year)+str(month)))
```

```python
# Function to calculate the Gaussian with constants a, b, and c
def gaussian(x, a, b, c):
    return a*np.exp(-np.power(x - b, 2)/(2*np.power(c, 2)))
```

```python
# Generate dummy dataset
x_dummy = np.array(p_numcontributors)
y_dummy = gaussian(x_dummy, 8, -1, 3)
```
